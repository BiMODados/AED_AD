# Importações Necessárias

```python
import pandas as pd
import numpy as np
from scipy.stats import kurtosis, skew
```

```python
df_cnaes = pd.read_excel("./cnaes/cnaes.xlsx")
#df_empresas7 = pd.read_excel('./Empresas7/empresas7.xlsx') # <- se lermos todos os arq de empresas passa do limite suportado pelo excel (dados em excesso)
df_empresas8 = pd.read_excel('./Empresas8/empresas8.xlsx')
df_empresas9 = pd.read_excel("./Empresas9/Empresas9.xlsx")
#df_estabelecimentos7 = pd.read_excel("./Estabelecimentos7/Estabelecimentos7.xlsx") # <- se lermos todos os arq de estabelecimentos passa do limite suportado upelo excel (dados em excesso)
df_estabelecimentos8 = pd.read_excel("./Estabelecimentos8/Estabelecimentos8.xlsx")
df_estabelecimentos9 = pd.read_excel("./Estabelecimentos9/Estabelecimentos9.xlsx")
df_motivos = pd.read_excel("./Motivos/motivos.xlsx")
df_municipios = pd.read_excel("./municipios/municipios.xlsx")
df_natureza_juridica = pd.read_excel("./nat_jud/natureza_juridica.xlsx")
df_pais = pd.read_excel("./pais/pais.xlsx")
df_qualificacao = pd.read_excel("./Qualificacoes/Qualificacoes.xlsx")
```

## Concatenando os dados referentes as empresas e estabelecimentos

```python
df_empresas = pd.concat([df_empresas8, df_empresas9])

df_estabelecimentos = pd.concat([df_estabelecimentos8, df_estabelecimentos9])
```

# Corrijindo os valores incorretos da coluna município da tabela estabelecimento

```python
# Converter valores não numéricos para NaN
df_estabelecimentos['Município'] = pd.to_numeric(df_estabelecimentos['Município'], errors='coerce')

# Remover linhas onde 'Municipio' é NaN
estabelecimentos = df_estabelecimentos.dropna(subset=['Município'])

# convertendo os valores de 'Municipio' de volta para inteiros
estabelecimentos['Município'] = estabelecimentos['Município'].astype(int)
```

# Combinando os datasets

```python
# Mostrando as colunas
print("Colunas Estabelecimentos:")
display(df_estabelecimentos.columns)
print("Colunas Cnaes:")
display(df_cnaes.columns)
print("Colunas Empresas:")
display(df_empresas.columns)
print("Colunas Motivos:")
display(df_motivos.columns)
print("Colunas Municipios:")
display(df_municipios.columns)
print("Colunas Natureza Juridica:")
display(df_natureza_juridica.columns)
print("Colunas Pais:")
display(df_pais.columns)
print("Colunas Qualificacoes:")
display(df_qualificacao.columns)


# 1. Estabelecimentos com Empresas (CNPJ Básico)
df_merged = pd.merge(df_estabelecimentos, df_empresas, on='CNPJ Básico', how='inner')

# 2. Estabelecimentos com Municipios (Município e id_mun)
df_merged = pd.merge(df_merged, df_municipios, left_on='Município', right_on='id_mun', how='inner')

# 3. Estabelecimentos com CNAEs (CNAE Fiscal Principal e id_cnaes)
df_merged = pd.merge(df_merged, df_cnaes, left_on='CNAE Fiscal Principal', right_on='id_cnaes', how='inner')

# 4. Estabelecimentos com Motivos (Motivo Situação Cadastral e id_mot)
df_merged = pd.merge(df_merged, df_motivos, left_on='Motivo Situação Cadastral', right_on='id_mot', how='inner')

# 5. Estabelecimentos com Natureza Jurídica (Natureza Jurídica e id_nat)
df_merged = pd.merge(df_merged, df_natureza_juridica[['id_nat', 'Natureza Juridica']], 
                     left_on='Natureza Jurídica', right_on='id_nat', how='inner')

# Exibindo o dataframe resultante
display(df_merged.head(2))

# Verificando as colunas
print("Colunas após o merge:")
print(df_merged.columns)


display(df_merged.head(2))
```

## Substituindo os códigos(valores int) pelos valores descritivos referentes a estes numeros, colunas afetadas:
- Situação Cadastral
- Porte Empresa

#### Situação Cadastral

```python
print('Dados diferentes na colunas Situação Cadastral: ',df_format["Situação Cadastral"].unique())

def mapear_situacao(situacao):
    if situacao in ['02', 2]:
        return 'NULA'
    elif situacao in ['08', 8]:
        return 'ATIVA'
    elif situacao in ['01', 1]:
        return 'SUSPENSA'
    elif situacao in ['04', 4]:
        return 'INAPTA'
    elif situacao in ['03', 3]:
        return 'BAIXADA'
    return situacao  # Retorna o valor original caso não seja nenhum dos mapeados


df_format['Situação Cadastral'] = df_format['Situação Cadastral'].apply(mapear_situacao)

print('Dados diferentes na colunas Situação Cadastral: ',df_format["Situação Cadastral"].unique())
```

## Porte da Empresa

```python
print('Dados diferentes na colunas Porte Empresa: ',df_format["Porte Empresa"].unique())

# Mapeamento dos códigos
porte_mapping = {
    0: 'Não Informado',
    1: 'Micro Empresa',
    3: 'Empresa de Pequeno Porte'
}

# Aplicar o mapeamento à coluna 'Porte Empresa'
df_format['Porte Empresa'] = df_format['Porte Empresa'].replace(porte_mapping)

# Para os demais valores que não estão no mapeamento
df_format['Porte Empresa'] = df_format['Porte Empresa'].where(df_format['Porte Empresa'].isin(porte_mapping.values()), 'Outros')

# Verificar os valores únicos após a substituição
print('Dados diferentes na colunas Porte Empresa: ',df_format["Porte Empresa"].unique())
```

## Normalizações

### Capping

```python
# Definindo os limites para o Capping nos dados
limite_superior = df_format['Capital Social'].quantile(0.994)  # 99,4%
limite_inferior = df_format['Capital Social'].quantile(0.006)  # 0,6%

# Criando o DataFrame para o Capping
df_capping = df_format.copy()
df_capping['Capital Social'] = df_capping['Capital Social'].clip(lower=limite_inferior, upper=limite_superior)

# Exibindo o DataFrame após o Capping
print("DataFrame com Capping aplicado:")
display(df_capping.sample(5))
```

### IQR

```python
# Função para calcular o IQR e remover outliers
def aplicar_iqr(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[coluna] >= lower_bound) & (df[coluna] <= upper_bound)]

# Criando o DataFrame para o IQR
df_iqr = aplicar_iqr(df_format.copy(), 'Capital Social')

# Exibindo o DataFrame após a aplicação do IQR
print("DataFrame com IQR aplicado:")
display(df_iqr.sample(5))
```

### Log

```python
# Criando o DataFrame para a Transformação Logarítmica
df_log = df_format.copy()

# Evitando valores negativos ou zero antes de aplicar a transformação logarítmica
df_log['Capital Social'] = df_log['Capital Social'].apply(lambda x: x if x > 0 else 1)
df_log['Capital Social'] = np.log(df_log['Capital Social'])

# Exibindo o DataFrame após a transformação logarítmica
print("DataFrame com Transformação Logarítmica aplicada:")
display(df_log.sample(5))
```

### Raiz

```python
# Criando o DataFrame para a Transformação de Raiz Quadrada
df_sqrt = df_format.copy()

# Aplicando a transformação de raiz quadrada
df_sqrt['Capital Social'] = np.sqrt(df_sqrt['Capital Social'])

# Exibindo o DataFrame após a transformação de raiz quadrada
print("DataFrame com Transformação de Raiz Quadrada aplicada:")
display(df_sqrt.sample(5))
```
